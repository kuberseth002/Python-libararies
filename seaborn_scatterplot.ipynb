```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a
url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

plt.figure(figsize=(8, 5))
sns.scatterplot(x="day", y="total_bill", data=df, hue="day", palette="viridis", s=100, alpha=0.7)

plt.title("Scatter Plot of Day vs Total Bill")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill (USD)")
plt.grid(True)

plt.show()
```
