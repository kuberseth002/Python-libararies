3. Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1

```python
x_values = []
y_values = []

with open("test.txt", "r") as file:
    for line in file:
        x, y = map(int, line.split())
        x_values.append(x)
        y_values.append(y)

plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label="Data Line")

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Line Plot from File Data")
plt.legend()

plt.show()
```

4. Write a Python program to draw line charts of the financial data of Alphabet Inc. between
October 3, 2016 to October 7, 2016.
Sample Financial data (fdata.csv):
Date,Open,High,Low,Close
03-10-16,774.25,776.065002,769.5,772.559998
04-10-16,776.030029,778.710022,772.890015,776.429993

06-10-16,779,780.47998,775.539978,776.859985

```python
df = pd.read_csv("fdata.csv")

df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%y")

# Step 3: Plot the financial data
plt.figure(figsize=(10, 5))

plt.plot(df['Date'], df['Open'], marker='o', ls='-', label="Open", color='blue')
plt.plot(df['Date'], df['High'], marker='o', ls='-', label="High", color='green')
plt.plot(df['Date'], df['Low'], marker='o', ls='-', label="Low", color='red')
plt.plot(df['Date'], df['Close'], marker='o', ls='-', label="Close", color='purple')

plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.title("Alphabet Inc. Financial Data (Oct 3 - Oct 7, 2016)")
plt.legend()
plt.grid()

# Step 5: Display the plot
plt.show()
```

12. Write a Python program to create multiple types of charts

```python
x = [0, 1, 2, 3, 4]
y1 = [50, 20, 30, 10, 40]
y2 = [40, 50, 20, 30, 10]
y3 = [30, 20, 50, 40, 10]
y4 = [40, 20, 10, 50, 30]

plt.figure(figsize=(10, 8)) 
plt.suptitle("Multiple types of charts", fontsize=14, fontweight='bold', color='darkblue')

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, marker='o', ls='-', c='blue', lw=2, markersize=8, label="Series 1")
plt.title("Plot 1: Line Chart", fontsize=12, c='darkred')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, marker='s', linestyle='--', color='green', linewidth=2, markersize=8, label="Series 2")
plt.title("Plot 2: Dashed Line", fontsize=12, color='darkred')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()

# Subplot 3
plt.subplot(2, 2, 3)
plt.scatter(x, y3, color='purple', s=80, edgecolors='black', label="Scatter Points")
plt.title("Plot 3: Scatter Plot", fontsize=12, color='darkred')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True, linestyle='-.', alpha=0.6)
plt.legend()

# Subplot 4
plt.subplot(2, 2, 4)
plt.bar(x, y4, color=['red', 'blue', 'green', 'orange', 'purple'], edgecolor='black', label="Bar Data")
plt.title("Plot 4: Bar Chart", fontsize=12, color='darkred')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend()

# Adjust layout to avoid overlap
plt.tight_layout(rect=[0, 0, 1, 0.95])  

# Show the plots
plt.show()
```

14. Write a Python program to display the grid and draw line charts of the closing value of
Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn
off ticks.

```python
data = {
    "Date": ["03-10-16", "04-10-16", "05-10-16", "06-10-16", "07-10-16"],
    "Close": [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
}

df = pd.DataFrame(data)

df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%y")

plt.figure(figsize=(10, 5))

plt.plot(df['Date'], df['Close'], marker='o', linestyle='-', color='blue', label="Closing Price")

plt.grid(True, which='minor', ls='-', lw=0.8, c='black') 
plt.grid(True, which='major', ls='--', lw=0.5, c='gray') 

#  Turn off ticks
plt.xticks([])
plt.yticks([])

plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.title("Alphabet Inc. Closing Prices (Oct 3 - Oct 7, 2016)")
plt.legend()

plt.show()
```
