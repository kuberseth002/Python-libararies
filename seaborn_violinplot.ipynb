```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Write a program to draw a violin plot of sex against total_bill for a given dataset
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

plt.figure(figsize=(8, 5))
sns.violinplot(x="sex", y="total_bill", data=df,hue="sex",palette="coolwarm")

plt.title("Violin Plot of Total Bill by Sex")
plt.ylabel("Total Bill (USD)")

plt.show()
```

2. Write a program to draw a violin plot of “species” against “sepal length” for a dataset
given in a url
https://github.com/mwaskom/seaborn-data/blob/master/iris.csv

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

plt.figure(figsize=(8, 5))
sns.violinplot(x="species", y="sepal_length", data=df, hue="species", palette="coolwarm", legend=False)

plt.title("Violin Plot of Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")

# Show the plot
plt.show()
```
