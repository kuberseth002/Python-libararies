```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Write a program to draw a point plot for sex against survived for a dataset given in url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
df = pd.read_csv(url)

plt.figure(figsize=(8, 5))
sns.pointplot(x="sex", y="survived", data=df, errorbar=None, marker="o", linestyle="solid", color="blue")

plt.title("Point Plot of Survival Rate by Sex on Titanic")
plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.ylim(0, 1)
plt.grid()

plt.show()
```
