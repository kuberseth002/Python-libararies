```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Write a Python program to draw a scatter graph taking a random distribution in X and Y and
plotted against each other.

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='blue', alpha=0.5)
plt.title("Random Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

2. Write a Python program to draw a scatter plot with empty circles taking a random distribution
in X and Y and plotted against each other

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, facecolors='none', edgecolors='blue')
plt.title("Scatter Plot with Empty Circles")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

3. Write a Python program to draw a scatter plot using random distributions to generate balls of
different sizes

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 1000  # Random sizes for scatter points

# Scatter plot
plt.scatter(x, y, s=sizes, color='green', alpha=0.5)
plt.title("Scatter Plot with Different Ball Sizes")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

4. Write a Python program to draw a scatter plot comparing two subject marks of Mathematics
and Science. Use marks of 10 students.
Test Data:
math_marks = [88, 92, 80, 89, 100, 80, 60, 100, 80, 34]

```python
math_marks = [88, 92, 80, 89, 100, 80, 60, 100, 80, 34]
science_marks = [35, 79, 79, 48, 100, 88, 32, 45, 20, 30]

plt.scatter(math_marks, science_marks, color='red', marker='o')
plt.title("Math vs Science Marks")
plt.xlabel("Math Marks")
plt.ylabel("Science Marks")
plt.grid(True)
plt.show()
```

5. Write a Python program to draw a scatter plot for three different groups camparing weights
and heights.

```python
group1_weight = np.random.randint(50, 80, 10)
group1_height = np.random.randint(150, 180, 10)

group2_weight = np.random.randint(60, 90, 10)
group2_height = np.random.randint(160, 190, 10)

group3_weight = np.random.randint(70, 100, 10)
group3_height = np.random.randint(170, 200, 10)

plt.scatter(group1_weight, group1_height, color='blue', label="Group 1")
plt.scatter(group2_weight, group2_height, color='green', label="Group 2")
plt.scatter(group3_weight, group3_height, color='red', label="Group 3")

plt.title("Height vs Weight Scatter Plot for 3 Groups")
plt.xlabel("Weight (kg)")
plt.ylabel("Height (cm)")
plt.legend()
plt.grid(True)
plt.show()
```
