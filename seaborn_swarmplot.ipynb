```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Write a program to draw a swarm plot of total bill against size for a given dataset
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

plt.figure(figsize=(10, 6))
sns.swarmplot(x="size", y="total_bill", data=df, hue="sex",palette="viridis", size=5)

plt.title("Swarm Plot of Total Bill vs. Size")
plt.xlabel("Table Size")
plt.ylabel("Total Bill (USD)")

plt.show()
```

2. Write a program to draw swarm plot of “total bill” against day for a dataset given in url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

plt.figure(figsize=(10, 6))
sns.swarmplot(x="day", y="total_bill", hue="sex", data=df, palette="coolwarm", size=5, dodge=True)

plt.title("Swarm Plot of Total Bill by Day (Grouped by Sex)")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill (USD)")
plt.legend(title="Sex")

plt.show()
```
